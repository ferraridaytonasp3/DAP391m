```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Tạo dữ liệu giả định
engine_cc = np.array([1400, 1600, 1800, 2000, 2200]) 
car_weight = np.array([580, 600, 630, 660, 690]) 

# Tạo dataframe
data = pd.DataFrame({
    'engine_cc': engine_cc,
    'car_weight': car_weight
})

data
```

```python
print(engine_cc)
X = engine_cc.reshape(-1,1)
y = car_weight

X
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print("Điểm cắt (intercept): " , model.intercept_)
print("Hệ số góc (slope): ", model.coef_[0])
print("R-squared:", model.score(X, y))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(engine_cc, car_weight, color='red', label='Dữ liệu gốc')
plt.plot(engine_cc, model.predict(X), color='blue', linewidth=2, label='Đường hồi quy tuyến tính')
plt.xlabel('Dung tích động cơ (cc)')
plt.ylabel('Khối lượng xe (kg)')
plt.title('Hồi quy tuyến tính cho khối lượng xe F1')
#plt.legend()
plt.show()
```

```python
new_engine_cc = np.array([1500, 1900, 2300]).reshape(-1,1)
predicted_weights = model.predict(new_engine_cc)

print("\nDự đoán khối lượng xe cho dung tích 1500cc, 1900cc và 2300cc:")
for cc, weight in zip(new_engine_cc.flatten(), predicted_weights):
   print(f"Dung tích {cc} cc: {weight:.2f} kg")
```
